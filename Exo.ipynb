```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('prets_final.csv')
df.head()
```

```python
# Sélection des lignes de prêts immobiliers et automobiles
immobilier = df[df["type"] == "immobilier"]
automobile = df[df["type"] == "automobile"]

# Calcul du nombre total de prêts immobiliers et automobiles
total_immobilier = len(immobilier)
total_automobile = len(automobile)
total_prets = total_immobilier + total_automobile

# Calcul de la proportion de prêts immobiliers et automobiles
proportion_immobilier = total_immobilier / total_prets
proportion_automobile = total_automobile / total_prets

# Tracé du graphique à secteurs
plt.pie([proportion_immobilier, proportion_automobile], labels=["immobilier", "automobile"], autopct="%1.1f%%")
plt.legend(title="Type de prêt")
plt.axis("equal")

# Affichage du graphique
plt.show()
```

```python
# Calcul du bénéfice mensuel par client
df["benefice"] = df["revenu"] * 0.1

# Tri des données par revenu croissant
df.sort_values("revenu", inplace=True)

plt.plot(df["revenu"], df["benefice"], color='red')
plt.xlabel("Revenu (euros)")
plt.ylabel("Bénéfice mensuel (euros)")
plt.title("Bénéfice mensuel en fonction du revenu")

# Affichage du graphique
plt.show()
```

```python
# Tracé de l'histogramme
plt.hist(df["benefice"], bins=20)
plt.xlabel("Bénéfice mensuel (euros)")
plt.ylabel("Nombre de clients")
plt.title("Distribution des bénéfices réalisés")

# Affichage du graphique
plt.show()
```

```python
agences = df.groupby("ville")["benefice"].sum()

# Tracé du graphique en bâtons
plt.bar(agences.index, agences.values)
plt.xlabel("Agence")
plt.ylabel("Bénéfice mensuel total (euros)")
plt.title("Bénéfice mensuel total par agence")

# Affichage du graphique
plt.show()
```
