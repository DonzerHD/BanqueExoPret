```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

prets = pd.read_csv("prets_final.csv")
prets.head()
```

```python
sns.countplot(x='type', data=prets)

# Ajout des labels aux axes
plt.xlabel("Type de prêt")
plt.ylabel("Nombre de prêts")

# Affichage du graphique
plt.show()
```

```python
immobilier = prets[prets['type'] == 'immobilier']

# Tracé du nuage de points
sns.scatterplot(x='revenu', y='benefices', hue='ville', data=immobilier)

# Ajout des labels aux axes
plt.xlabel("Revenu (euros)")
plt.ylabel("Bénéfice mensuel (euros)")

# Affichage du graphique
plt.show()
```

```python
# Tracé de l'histogramme
sns.histplot(data=prets, x='benefices')

# Ajout des labels aux axes
plt.xlabel("Bénéfice mensuel (euros)")
plt.ylabel("Nombre de prêts")

# Affichage du graphique
plt.show()

# Tracé du diagramme à densité
sns.kdeplot(data=prets, x='benefices')

# Ajout des labels aux axes
plt.xlabel("Bénéfice mensuel (euros)")
plt.ylabel("Densité de probabilité")

# Affichage du graphique
plt.show()
```

```python
# Agrégation du bénéfice par agence et type de prêt
agregate = prets.groupby(['ville', 'type'])['benefices'].sum().reset_index()

# Tracé du diagramme en barres
sns.barplot(x='ville', y='benefices', hue='type', data=agregate)

# Ajout des labels aux axes
plt.xlabel("ville")
plt.ylabel("Bénéfice mensuel total (euros)")

# Affichage du graphique
plt.show()
```
